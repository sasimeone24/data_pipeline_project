```python
import numpy as np
import pandas as pd

import time
import datetime
import random
import math
import re
import progressbar

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def random_move():
    
    """
    This function is in charge of choosing a random value
    from the keys list and then returning it to be used by Selenium.
    
    Return
        action (Random element from the keys list)
    """
    
    keys = [Keys.ARROW_DOWN, Keys.ARROW_UP, Keys.END, Keys.PAGE_UP, Keys.ARROW_LEFT, Keys.ARROW_RIGHT]
    action = np.random.choice(keys, p=[0.3, 0.3, 0.1, 0.1, 0.1, 0.1])
    
    return action
```

```python
def get_links(driver, url, pages):
    
    """
    We can obtain the HTML code from the Metacritics page 
    where the links to the reviews of the games are found.
    
    Return
        HTML code
    """
    
    print('The process to get the links has started\n')
    
    source_code = []
    for i in range(pages):
        driver.get(url % i)
        
        time.sleep(2)
        ActionChains(driver).send_keys(random_move()).perform()  # Le pedimos que mueva la pagina de forma aleatoria
        
        source_code.append(driver.page_source)
        time.sleep(np.random.randint(1, high=8)) # Espera un tiempo aleatorio antes de pasar a la siguiente pagina
    
    return source_code
```

```python
def usable_links(source_code):
    
    """
    It obtains the links of each of the games and 
    we save it in a list.
    
    Return
        links_correct
    """
    
    base_url = 'https://www.metacritic.com'
    links_correct =[]

    for page in range(len(source_code)):
        soup = BeautifulSoup(source_code[page], 'html.parser')
        link = soup.select('td[class="clamp-summary-wrap"] > a[class="title"]')
        links_correct.append([base_url + i['href'] for i in link])
        
    links_correct = list(np.concatenate(links_correct).flat) # Concatenamos la distintas listas y la volvemos plana
    print('Game links have been scraped\n')
    
    return links_correct
```

```python
def get_info(driver, num_games, links):
    
    """
    Get the HTML code for each game page. Call the 
    info_dataframe function to get the desired information.
    
    Return
        Dataframe whit the games information
    """
    
    print('The process to get the game data has started\n')
    
    columns = ['Name', 'Release_date', 'Players', 'Genre', 'Metascore', 'Rating_Metascore', # Nombres de las columnas 
              'User_score', 'Rating_User_score', 'Awards', 'Distributor', 'Developer', 
              'Rating', 'Consoles', 'Summary'] 
    df_game = pd.DataFrame(columns=columns)
    count = 0
    i = 0
    
    bar = progressbar.ProgressBar(maxval=num_games, widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
    bar.start()
    
    while count != num_games:
        driver.get(links[i])
        time.sleep(2)
        
        ActionChains(driver).send_keys(random_move()).perform()
    
        source_code = driver.page_source
        time.sleep(np.random.randint(1, high=8))
        
        soup = BeautifulSoup(source_code, 'html.parser')
        # Un try and except para prevenir un error en caso de que la pagina no cargue
        try: 
            # Si el nombre del juego ya esta en la lista pasamos al siguiente
            if soup.select('div.product_title > a')[0].text.strip() not in list(df_game['Name']):
                df = info_dataframe(soup)
                df.columns = columns
                df_game = pd.concat([df_game, df])
                count += 1
                bar.update(count)
    
        except IndexError: # Si la pagina no carga tendremos una lista vacia. En ese caso tendremos un IndexError
            name = ' '.join(re.findall(r'[\w+\-]*[\w+]$', links[i])[0].split('-'))
            print(f'The page for the game: {name} \ndid not load properly and the information could not be obtained')
        
        i += 1
    
    bar.finish()
    print('Game data has been scraped\n')
    return df_game
```

```python
def nan_values(category):
    
    """
    Detects if a field is empty and returns NaN in that case.
    Otherwise it returns the text of the input parameter without line breaks.
    
    Return
        NaN if the field is empty
        Text without line breaks if the field is not empty
    """
    
    if len(category) != 0:
        if len(category) == 1:
            return [i.text.strip() for i in category][0]
        if len(category) > 1:
            return [i.text.strip() for i in category]
    
    else:
        return math.nan
```

```python
def info_dataframe(soup_game):
    
    """
    We search and extract the desired information 
    from each game page through HTML code and BeautifulSoup.
    
    Return
        DataFrame
    """
    
    metascore = soup_game.select('a[class="metascore_anchor"] > div > span') 
    metascore_text = nan_values(metascore)
    
    user_score = soup_game.select('a[class="metascore_anchor"] > div')
    user_score_text = nan_values(user_score)[1]

    name = soup_game.select('div.product_title > a')
    name_text = nan_values(name)
    
    release_date = soup_game.select('li[class="summary_detail release_data"] > span[class="data"]') 
    release_date_text = nan_values(release_date)
    
    rating = soup_game.select('li[class="summary_detail product_rating"] > span[class="data"]')
    rating_text = nan_values(rating)
    
    developer = soup_game.select('li[class="summary_detail developer"] > span[class="data"]')
    developer_text = nan_values(developer)
    
    summary = soup_game.select('li[class="summary_detail product_summary"] > span[class="data"]')
    summary_text = nan_values(summary)
    
    distributor = soup_game.select('li[class="summary_detail publisher"] > span[class="data"]')
    distributor_text = nan_values(distributor)
    
    genre = soup_game.select('li[class="summary_detail product_genre"] > span[class="data"]')
    genre_text = nan_values(genre)
    
    number_players = soup_game.select('li[class="summary_detail product_players"] > span[class="data"]')
    number_players_text = nan_values(number_players)
    
    awards = soup_game.select('td[class="ranking_wrap"] > div > a')
    awards_text = nan_values(awards)
    
    consoles = soup_game.select('span[class="platform"], li[class="summary_detail product_platforms"] >\
                                        span[class="data"] > a')
    consoles_text = nan_values(consoles)
    
    ratings_user_score = soup_game.select('div[class="userscore_wrap feature_userscore"] > div[class="summary"] > p >\
                                        span[class="count"] > a')
    ratings_user_score_text = nan_values(ratings_user_score)

    ratings_metascore = soup_game.select('div[class="summary"] > p > span > a > span')
    ratings_metascore_text = nan_values(ratings_metascore)
    
 
    df = pd.DataFrame([[name_text, release_date_text, number_players_text, genre_text, metascore_text,
                        ratings_metascore_text, user_score_text, ratings_user_score_text, awards_text, 
                        distributor_text, developer_text, rating_text, consoles_text, summary_text]]
                     )
    
    return df
```

```python
def avoid_duplicate(lista):
    
    """
    In the list we can find the same game but from different 
    consoles that internally contain the same information. 
    To avoid this, we check if the name is already in the list; 
    in case it is true we do not take the information of that 
    game and we go to the next one.
    """
    
    genre = []
    if isinstance(lista, list):
        for i in lista:
            if i not in genre:
                genre.append(i)
        return ', '.join(genre)
    
    else:
        return lista
```

```python
def clean_table(df):
    
    """
    We pass the df to it and clean each of the columns. 
    There are categorical type columns that must be numeric, 
    and we have a column that we have to change to datetime.
    
    Return
        DataFrame
    """

    df['Metascore'] = df['Metascore'].apply(int)
    df['Rating_Metascore'] = df['Rating_Metascore'].apply(int)
    df['User_score'] = df['User_score'].apply(float)
    df['Rating_User_score'] = df['Rating_User_score'].apply(lambda x: int(x.split()[0]))
    df['Release_date'] = df['Release_date'].apply(lambda x: datetime.datetime.strptime(x, '%b %d, %Y'))
    df['Awards'] = df['Awards'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
    df['Consoles'] = df['Consoles'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
    df['Summary'] = df['Summary'].apply(lambda x: x[0:-8]).apply(lambda x: x.replace('\n', ''))
    df['Distributor'] = df['Distributor'].apply(lambda x: ' '.join(x.split()))
    df['Developer'] = df['Developer'].apply(lambda x: ' '.join(x.split()))
    df['Genre'] = df['Genre'].apply(avoid_duplicate)

    return df
```

```python
def filters(df_clean):
    
    """
    This function will be in charge of analyzing the data. 
    Create filters for later use
    
    Return
        Filters
    """
    
    df = df_clean.copy()
    
    # Obtenemos el Distribuidor con mas juegos
    top_distr = df.groupby('Distributor').agg(count_games=('Distributor', 'count')).sort_values('count_games', ascending=False)
    
    # Obtenemos el Desarrolador con mas juegos
    top_dev = df.groupby('Developer').agg(count_games=('Developer', 'count')).sort_values('count_games', ascending=False)
    
    # Obtenemos la cuenta de los juegos por Rating
    rating_count = df.groupby('Rating').agg(number=('Genre', 'count'))
    
    # Obtenemos el numero de juegos en la lista por años
    df['Release_date'] = pd.to_datetime(df['Release_date'], format='%Y-%m-%d',) # El formato lo tenemos que cambiar
    df['Year'] = df['Release_date'].dt.year # Creamos un lista de años
    year_count_games = df.groupby('Year').agg(games_year=('Name', 'count'))
    
    # Obtenemos el numero de juegos disponibles en el top para cada consola
    console_list = []
    for i in df['Consoles']:
        for j in i.split(','):
            word = re.findall(r'\w.+', j)
            if word[0] not in console_list:
                console_list.append(word[0]) # Obtenemos la lista de consolas

    count_console = [] 
    count = 0
    for i in console_list: # Agarramos cada elemento de la lista
        for j in df['Consoles']: # Tomamos cada string de nuestro datos
            for k in j.split(', '): # El string lo separamos en una lista y tomamos cada valor
                if i == k: # Si tanto el string de la lista como el obtenido con el split son iguales es porque es la misma consola
                    count += 1           
        count_console.append(count)
        count = 0
    games_console = pd.DataFrame([count_console], columns=console_list).transpose().sort_values(0,ascending=False).head(10)
    
    # Obtenemos la cuenta del top 10 de generos de todos los juegos 
    genre_list = []
    for i in df['Genre']:
        for j in i.replace(',', '').replace('/', '').split():
            if j not in genre_list:
                genre_list.append(j)         
    count_genre = []   
    for i in genre_list:
        count = df[df['Genre'].str.contains(i)].count().iloc[3]
        count_genre.append(count)
    games_genre = pd.DataFrame([count_genre], columns=genre_list).transpose().sort_values(0,ascending=False).head(10)
        
    return top_distr, top_dev, year_count_games, games_console, games_genre, rating_count
```

```python
def visualization(top_distr, top_dev, year_count_games, games_console, games_genre, rating_count):
    
    """
    It is in charge of showing the filters created 
    through graphs.
    """
    
    # Grafica de barras para el dataframe top_distr
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=top_distr.transpose().iloc[:, :10])
    ax.set(xlabel='Distributor Name', ylabel='Count of Games')
    plt.title('Top 10 Distributors for Best Rating Games in Metacritics' + "\n", fontsize=16)
    
    # Grafica de barras para el dataframe top_dev
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=top_dev.transpose().iloc[:, :10])
    ax.set(xlabel='Developer Name', ylabel='Count of Games')
    plt.title('Top 10 Developer for Best Rating Games in Metacritics' + "\n", fontsize=16)
    
    # Grafica de barras para el dataframe year_count_games
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=year_count_games.transpose())
    ax.set(xlabel='Year', ylabel='Count of Games')
    plt.title('Number of Games by Year in de top Rating games' + "\n", fontsize=16)

    # Grafica de barras para el dataframe games_genre
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=games_genre.transpose().iloc[:, :10])
    ax.set(xlabel='Genre', ylabel='Count of Games')
    plt.title('Top 10 Genre of Games' + "\n", fontsize=16)

    # Grafica de barras para el dataframe games_console
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=games_console.transpose().iloc[:, :10])
    ax.set(xlabel='Consoles', ylabel='Count of Games')
    plt.title('Top 10 consoles with the most games in the ranking' + "\n", fontsize=16) 
    
    # Grafica circular de los ratings
    colors = sns.color_palette('bright')[0:5]
    fig, ax = plt.subplots(figsize=(25,10))
    plt.pie(rating_count.iloc[:, 0], labels = rating_count.index, colors = colors, autopct='%.0f%%')
    plt.title('Percentage of Ratings' + "\n", fontsize=16)
    plt.show()
```

```python
def main(num_pages_games, num_games):
    
    """
    This is the main function where the other 
    functions of our code are housed
    """
    print('\nProgram started!\n')
    
    url = 'https://www.metacritic.com/browse/games/score/metascore/all/all/filtered?page=%s'
    path = r'C:\Program Files\ChromeDriver\chromedriver.exe'

    options = webdriver.ChromeOptions()
    options.add_argument('--incognito')
    options.add_argument('--headless')

    driver = webdriver.Chrome(path, options=options) # Iniciamos el Driver
    source_code = get_links(driver, url, num_pages_games)
    links = usable_links(source_code) # Lista de links con nuestras paginas
 
    df = get_info(driver, num_games, links)
    df.reset_index(drop=True, inplace=True) # Reiniciamos los index de nuestro dataframe

    driver.quit()

    df_clean = clean_table(df)
    df_clean.to_csv('Data_sets\metacritics_clean.csv', sep=',', index=False) # Guardamos la data en un CSV
    
    print('-----' * 5)
    print('The data has been imported to a csv the path is: Data_sets\metacritics_clean.csv')
    print('-----' * 5)
    
    # Analizamos nuestros datos
    top_distr, top_dev, year_count_games, games_console, games_genre, rating_count = filters(df_clean)
    
    # Procemos a la etapa de visualizar los datos
    visualization(top_distr, top_dev, year_count_games, games_console, games_genre, rating_count)
    
    print('Program has ended!\n')
```

```python
# Desde aqui corremos todo nuestro codigo.
i = False
while i == False:
    try:
        num_pages_games = int(input('Enter the number of pages to scrape where the links are: \n')) # Defoultd 6
        num_games = int(input('Enter the number of unique games to scrape: \n')) # Defoultd 300
        i = True
        
    except ValueError:
        print('Enter whole numbers\n')

main(num_pages_games, num_games)
```
